**Data Cleaning**

```python
df.drop(['URL','Full_Address','PhoneNumber','KnownFor','Dinner Reviews','Delivery Reviews','Timing'],axis=1,inplace=True)
df.head()
```

```python
#Replacing the '-' to '0' from the Dinner Ratings column
df['Dinner Ratings'] = df['Dinner Ratings'].replace('-','0')
df['Dinner Ratings']
```

```python
#Replacing the '-' to '0' from the Delivery Ratings column
df['Delivery Ratings'] = df['Delivery Ratings'].replace('-','0')
df['Delivery Ratings']
```

```python
#Converting the datatype of Dinner Ratings column from object to float.
df['Dinner Ratings'] = df['Dinner Ratings'].astype('float64')
```

```python
#Converting the datatype of Delivery Ratings column from object to float.
df['Delivery Ratings'] = df['Delivery Ratings'].astype('float64')
```

```python
#Filling the missing data present in the 'PopularDishes' column.
df['PopularDishes']=df['PopularDishes'].fillna('No Data')
```

```python
#Filling the missing data present in the 'PeopleKnownFor' column.
df['PeopleKnownFor']=df['PeopleKnownFor'].fillna('No Data')
```

```python
df.info()
```

# Summary

Removed URL, Full_Address, PhoneNumber, KnownFor, Dinner Reviews, Delivery Reviews, Timing columns from the dataset since it won't be required in this analysis.

Replaced '-' to '0' from Dinner Ratings and Delivery Ratings column to convert the datatype from object to float so that we can perform numerical calculations on it.

Filled the missing data present in PopularDishes and PeopleKnownFor column with 'No Data' to remove all the missing values and help in better analysis.

After performing these steps our dataset is finally clean, having no null values or messy data. Now, let's jump into the visualisation part.

10. Top ten areas having highest ratings for deliveries.

```python
top10_area = df.groupby('Area')['Delivery Ratings'].mean().sort_values(ascending=False).head(10)

sns.barplot(x=top10_area.index, y=top10_area.values)
plt.title("Area-wise Delivery Ratings")
plt.xlabel("Areas")
plt.ylabel("Delivery Ratings")
plt.xticks(rotation=90)
plt.show()
```
